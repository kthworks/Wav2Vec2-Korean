```python
from transformers import Wav2Vec2CTCTokenizer, Wav2Vec2FeatureExtractor, Wav2Vec2Processor,  Wav2Vec2ForCTC
processor = Wav2Vec2Processor.from_pretrained("Taeham/wav2vec2-ksponspeech-dataset")
model = Wav2Vec2ForCTC.from_pretrained("Taeham/wav2vec2-ksponspeech-dataset").to('cuda')
from datasets import load_dataset
ds = load_dataset("Taeham/wav2vec2-ksponspeech-third-party-test")
test = ds['train']
test
```

```python
import torch
import librosa
from matplotlib import pyplot as plt
import librosa.display
import IPython.display as ipd

sample = 33

tmp = processor(test['audio'][sample], sampling_rate = 16000, return_tensors='pt', padding=True)
input_values = tmp.input_values.to('cuda')
with torch.no_grad():
    logits = model(input_values.type(torch.cuda.FloatTensor)).logits
predicted_ids = torch.argmax(logits, dim=-1)
pred = processor.batch_decode(predicted_ids)


from jaso import jaso
print(['ANSWER : ', jaso().to_sentence(test['jaso_text'][sample])])
print(['decoded : ', jaso().to_sentence(''.join(pred))])

ipd.Audio(test['audio'][sample], rate=16000)
```
